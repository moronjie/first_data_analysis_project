```python
df2 = df[((df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2022)) & (df['Event name'].str.contains('USA'))]
df2
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
df2
```

```python
df2['athlete_age'] = 2022 - df2['Athlete year of birth']
df2
```

```python
# drop some columns 
df2.drop(['Athlete club', 'Athlete year of birth', 'Athlete age category'], axis=1, inplace=True)
df2
```

```python
df2.reset_index(drop=True, inplace=True)
df2
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(" ").str.get(0)
df2
```

```python
df2.isnull().sum()
```

```python
df2 = df2.dropna()
df2
```

```python
df2.info()
```

```python
df2['athlete_age'] = df2['athlete_age'].astype('int')
df2['Athlete average speed'] = df2['Athlete average speed'].astype('float')
df2
```

```python
df2[df2.duplicated()]
```

```python
df2 = df2.rename(columns={
    'Year of event': 'year',
    'Event dates': 'date',
    'Event name': 'event_name',
    'Event distance/length': 'distance',
    'Event number of finishers': 'number_of_fiishers',
    'Athlete performance': 'athlete_performance',
    'Athlete country': 'country',
    'Athlete gender': 'gender',
    'Athlete average speed': 'average_speed',
    'Athlete ID': 'athlete_id'
})
df2
```

difference in speed average for gender in each distance. 50km always had the higher number of atheletes since it has the sorter distance 

```python
# diff in speed 
df2.groupby(['distance', 'gender'])['average_speed'].mean()
```

the best age in 50mi where age is greater than 20


```python
# age with best 50mi race 
df2.query('distance == "50mi"').groupby('athlete_age')['average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count > 20')
```

the worse age in 50mi where age is greater than 20


```python
df2.query('distance == "50mi"').groupby('athlete_age')['average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count > 20')
```
